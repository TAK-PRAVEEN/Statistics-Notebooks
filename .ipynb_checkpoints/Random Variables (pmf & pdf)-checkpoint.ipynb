# Random Variable
    @ A random variable is a variable whose value is a numerical outcome of a random phenomenon. It links outcomes of a random process to numerical values. There are two main types:

* Discrete Random Variable: Takes on a finite or countably infinite number of values (e.g., the result of a dice roll).

* Continuous Random Variable: Takes on any value within a given range or interval (e.g., the height of a person).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import (uniform, norm, bernoulli, binom, poisson, geom, expon, gamma, chi2, t, f, lognorm, pareto)

import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (8, 5)
```

```python
def plot_distribution(x, y, title, kind='plot', xlabel='x', ylabel='Probability/Density'):
    plt.figure()
    if kind == 'bar':
        plt.bar(x, y, color='skyblue', edgecolor='black')
    else:
        plt.plot(x, y, color='darkred')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()
```

--- 
## 1. Probability Mass Function (PMF)
> The Probability Mass Function (PMF), denoted as P(X=x), gives the probability that a discrete random variable X is exactly equal to some value x. The sum of all probabilities in a PMF is always 1.

### 1.1 Discrete Uniform Distribution: 
    Models a situation where all outcomes are equally likely.
+ Description: A variable X has k possible outcomes, each with the same probability.
+ PMF: $ P(X=x)= \frac{1}{k} $
+ Parameters: $ k $ (number of possible outcomes).
+ Example: A single roll of a fair 6-sided die, where $ k $=6 and the probability for each outcome {1, 2, 3, 4, 5, 6} is 1/6.

```python
x = np.arange(1, 7)
y = np.ones_like(x) / len(x)
plot_distribution(x, y, "Discrete Uniform Distribution (Die Roll)", kind="bar")
```

### 1.2 Bernoulli Distribution:
    The simplest case of a binary outcome.
+ Description: A single trial with only two possible outcomes: success (1) or failure (0).
+ PMF: $ P(X=x) = p^{x}(1-p)^{1-x} $ for x ∈ {0,1}
+ Parameters: $ p $ (probability of success).
+ Example: A single coin flip, where success (Heads) has probability $ p $.

```python
x = [0, 1]
y = bernoulli.pmf(x, p=0.6)
plot_distribution(x, y, "bernoulli Distribution (p=0.6)", kind="bar")
```

### 1.3 Binomial Distribution
    Models the number of successes in a fixed number of independent trials.
+ Description: The number of successes, k, in n independent Bernoulli trials.
+ PMF: $ P(X=k) = \binom{n}{r}p^{k}(1-p)^{n-k} $
+ Parameters: $ n $ (number of trials), $ p $ (probability of success in one trial).
+ Example: The number of heads in 10 coin flips.

```python
x = np.arange(0, 11)
y = binom.pmf(x, n=10, p=0.5)
plot_distribution(x, y, "Binomial Distribution (n=10, p=0.5)", kind="bar")
```

### 1.4 Poisson Distribution
    Models the number of events occurring in a fixed interval of time or space.
+ Description: The number of events, k, happening in a fixed interval, given the average rate of occurrence.
+ PMF: $ P(X=k) = \frac{λ^{k}e^{-λ}}{k!} $
+ Parameters: λ (lambda), the average number of events per interval.
+ Example: The number of calls received by a call center in one hour.

```python
x = np.arange(0, 15)
y = poisson.pmf(x, mu=3)
plot_distribution(x, y, "Poisson Distribution (λ=3)", kind="bar")
```

### 1.5 Geometric Distribution
    Models the number of trials needed to get the first success.
+ Description: The number of trials, k, up to and including the first success in a series of independent Bernoulli trials.
+ PMF: $ P(X=k) = (1−p)^{k-1}p $
+ Parameters: $ p $ (probability of success in one trial).
+ Example: The number of coin flips needed to get the first head.

```python
x = np.arange(1, 15)
y = geom.pmf(x, p=0.3)
plot_distribution(x, y, "Geometric Distribution (p=0.3)", kind="bar")
```

## 2. Probability Density Function (PDF)
> The Probability Density Function (PDF), denoted as $ f(x) $, is used for continuous random variables. The probability of the variable falling within a particular range is given by the integral (area under the curve) of the PDF over that range. The value $ f(x) $ itself is not a probability; the total area under the entire curve is always 1.

### 2.1 Continuous Uniform Distribution
    Models a situation where any value within a range is equally likely.
+ Description: All values in a range [a,b] have an equal chance of occurring.
+ PDF: $ f(x)= \frac{1}{b-a} $ for $ x ∈ [a,b] $, and 0 otherwise.
+ Parameters: $ a $ (minimum value), $ b $ (maximum value).

```python
x = np.linspace(0, 1, 100)
y = uniform.pdf(x, loc=0, scale=1)
plot_distribution(x, y, "Continuous Uniform Distribution (0 to 1)")
```

### 2.2 Normal Distribution (Gaussian)
    The most famous distribution, characterized by its bell shape.
+ Description: Describes many natural phenomena like height, blood pressure, etc. It's symmetric around its mean.
+ PDF: $ f(x)= \frac{1}{σ\sqrt{2π}}e^{-\frac{1}{2}(\frac{x−μ}{σ})^{2}} $
+ Parameters: $ μ $ (mu), the mean; $ σ^2 $ (sigma-squared), the variance.

```python
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, loc=0, scale=1)
plot_distribution(x, y, "Normal Distribution (μ=0, σ=1)")
```

### 2.3 Standard Normal Distribution
    A special case of the normal distribution, used for standardization.
+ Description: A normal distribution with a mean of 0 and a standard deviation (and variance) of 1. A random variable $ X∼N(μ,σ ^2) $ can be standardized to $ Z=\frac{X−μ}{σ} $.
+ PDF: $ f(z)= \frac{1}{\sqrt{2π}}e^{-\frac{z^{2}}{2}} $
+ Parameters: $ μ $=0, $ σ^2 $=1.

```python
x =  np.linspace(-4, 4, 1000)
y = norm.pdf(x)
plot_distribution(x, y, "Standard Normal Distribution")
```

### 2.4 Lognormal Distribution
    A distribution of a random variable whose logarithm is normally distributed.
+ Description: Used for variables that are positively skewed, like income or stock prices.
+ PDF: $ f(x)= \frac{1}{xσ\sqrt{2π}}e^{-\frac{(ln(x)−μ)^{2}}{2σ^{2}}} $ for $ x>0 $.
+ Parameters:$ μ $ and $ σ $ of the variable's natural logarithm.

```python
x = np.linspace(0.01, 5, 1000)
y = lognorm.pdf(x, s=0.5)
plot_distribution(x, y, "Log-Normal Distribution (σ=0.5)")
```

### 2.5 Exponential Distribution
    Models the time until an event occurs.
+ Description: The time between events in a Poisson process. It's memoryless.
+ PDF: $ f(x) = λe^{-λx} $ for $ x≥0 $.
+ Parameters: λ (lambda), the rate parameter.

```python
x = np.linspace(0, 10, 1000)
y = expon.pdf(x, scale=1)
plot_distribution(x, y, "Exponential Distribution (λ=1)")
```

### 2.6 Gamma Distribution
    A flexible distribution that generalizes the exponential and chi-squared distributions.
+ Description: Models the waiting time until the k-th event in a Poisson process.
+ PDF: $ f(x) = \frac{λ^{k}x^{k-1}e^{−λx}}{\Gamma{(k)}} $ for $ x≥0 $.
+ Parameters: $ k $ (shape), $ λ $ (rate). $ \Gamma{(k)} $ is the gamma function.

```python
x = np.linspace(0, 20, 1000)
y = gamma.pdf(x, a=2, scale=2)
plot_distribution(x, y, "Gamma Distribution (α=2, θ=2)")
```

### 2.7 Chi-Squared ($ χ^2 $) Distribution
+ Description: The distribution of a sum of squared standard normal deviates. Crucial for goodness-of-fit tests and confidence intervals for variance.
+ PDF: The PDF is complex. It's a special case of the Gamma distribution. $$ f(x,k) = \frac{1}{2^{k/2} \Gamma{(k/2)}}x^{(k/2)-1}e^{-x/2}, x ≥ 0 $$
+ Parameters: k (degrees of freedom), which is the number of squared normal deviates being summed.ng summed.

```python
x = np.linspace(0, 20, 1000)
y = chi2.pdf(x,df=5)
plot_distribution(x, y, "Chi-Sqaure Distribution (df=5)")
```

### 2.8 Student's t-Distribution
+ Description: Similar to the normal distribution but with heavier tails. Used for estimating the mean of a normally distributed population when the sample size is small and the population standard deviation is unknown.
+ Parameters: $ \nu $(nu), the degrees of freedom. As $ \nu $→∞, the t-distribution approaches the standard normal distribution.

```python
x = np.linspace(-5, 5, 1000)
y = t.pdf(x, df=10)
plot_distribution(x, y, "t-Distribution (df=10)")
```

### 2.9 F-Distribution
+ Description: The distribution of the ratio of two independent chi-squared variables divided by their respective degrees of freedom. Used in ANOVA to compare the means of multiple groups.
+ Parameters: $ d_1 $ and $ d_2 $, the degrees of freedom for the two chi-squared variables.

```python
x = np.linspace(0.01, 5, 1000)
y = f.pdf(x, dfn=5, dfd=10)
plot_distribution(x, y, "f-Distribution (df1=5, df2=10)")
```

### 2.10 Pareto Distribution
+ Description: Follows the "80-20 rule," where roughly 80% of the effects come from 20% of the causes. Used to model wealth distribution, city populations, etc.
+ PDF: $ f(x)= \frac{α{x_m}^{α}}{x^{α+1}} $ for $ x≥x_m $.
+ Parameters: $ x_m $ (scale, the minimum possible value), $ α $ (shape).

```python
x = np.linspace(1, 10, 1000)
y = pareto.pdf(x, b=3)
plot_distribution(x, y, "Pareto Distribution (α=3)")
```
